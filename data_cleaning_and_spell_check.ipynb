```python
import pandas as pd

# Read CSV file into DataFrame df
df = pd.read_csv('text_emotion.csv', encoding='utf-8')

# Show dataframe
print(df)
```

```python
df['sentiment'].value_counts()
```

```python
tweets = df['content']
```

```python
#https://www.analyticsvidhya.com/blog/2014/11/text-data-cleaning-steps-python/

import itertools
#Remove all XML/HTML tags, like &quot;
#Replace all instances of 's, 're et cetera to correct spelling
tweets_no_HTML = []
for sentence in range(len(tweets)):
    current = tweets[sentence]
    HTML_Appostophes = ["&amp;", " and ", "&quot;", "''", "&lt;", " < ", "&gt", "> ", "'s", " is", "'re", " are", "i", "I"]
    for i in range(len(HTML_Appostophes)-1):
        current = current.replace(HTML_Appostophes[i], HTML_Appostophes[i+1])
        
    words = ''.join(''.join(s)[:2] for _, s in itertools.groupby(current))
    tweets_no_HTML.append(words)
```

```python
tweets_update = []
for tweet in tweets_no_HTML:
    tweet = tweet.split()
    new_sentence = []
    for word in tweet:
        if word[0] != '@':
            new_sentence.append(word)
    
    new_sentence = " ".join(new_sentence)
    tweets_update.append(new_sentence)

for tweet in tweets_update:
    print(tweet)
    
        
```

```python
#https://www.geeksforgeeks.org/spelling-checker-in-python/
#https://pypi.org/project/pyspellchecker/    
from spellchecker import SpellChecker
spell = SpellChecker()

txt = tweets_update[5]
txt = txt.split()
new_sentence = []


for word in txt:
    if str(word) in spell:
        new_sentence.append(word)
    else:
        new_sentence.append(spell.correction(str(word)))
    
print(txt)
print(new_sentence)
```
